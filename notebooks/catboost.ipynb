```python
import pyarrow.parquet as pq
from pandas import DataFrame


def load_data_next_batch(path, batch_size=10_000):
    """Read a singular batch from .parquet file"""
    parquet_file = pq.ParquetFile(path)

    iterator = parquet_file.iter_batches(batch_size)
    next(iterator)
    batch = next(iterator)
    df: DataFrame = batch.to_pandas()
    return df
```

```python
df = load_data_next_batch("../data/train/train_data.parquet", batch_size).drop(['variantid1', 'variantid2'], axis=1)

# Spread embedding2
expanded_columns = pd.DataFrame(df['embedding2'].tolist(), index=df.index)
df = pd.concat([df.drop('embedding2', axis=1), expanded_columns], axis=1)

# Rename columns
new_columns = {i: i + 320 for i in range(320)}
df.rename(columns=new_columns, inplace=True)

# Spread embedding1
expanded_columns = pd.DataFrame(df['embedding1'].tolist(), index=df.index)
df = pd.concat([df.drop('embedding1', axis=1), expanded_columns], axis=1)
```

```python
X_test = df.drop(["target"], axis=1)
y_test = df["target"]
```

```python
from sklearn.metrics import f1_score, accuracy_score


y_pred = model.predict(X_test)

print("F1:", f1_score(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```
