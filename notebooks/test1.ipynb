```python
import pandas as pd
pd.set_option('display.max_colwidth', 300)
# Загрузка файла в DataFrame
df_a = pd.read_parquet('data/train/attributes.parquet')
df_res = pd.read_parquet('data/train/resnet.parquet')
df_tab = pd.read_parquet('data/train/text_and_bert.parquet')
df_train = pd.read_parquet('data/train/train.parquet')

# Просмотр первых нескольких строк
#df_train.head()
```

```python
def get_full_info_by_id(id, df_a, df_tab, df_res):
    a = df_a[df_a['variantid'] == id].iloc[0].to_dict()
    tab = df_tab[df_tab['variantid'] == id].iloc[0].to_dict()
    res = df_res[df_res['variantid'] == id].iloc[0].to_dict()
    full_info = {}
    full_info.update(a)
    full_info.update(tab)
    full_info.update(res)
    
    return full_info
```
